# Importando as Bases de dados e salvando em váriaveis

```python
import pandas as pd

#TABELA 2020 PARTE 1
tabela2020_p1 = pd.read_csv('HIST_PAINEL_COVIDBR_2020_Parte1_24jun2022.csv',sep=';', decimal=',')
#TABELA 2020 PARTE 2
tabela2020_p2 = pd.read_csv('HIST_PAINEL_COVIDBR_2020_Parte2_24jun2022.csv',sep=';', decimal=',')
#TABELA 2021 PARTE 1
tabela2021_p1 = pd.read_csv('HIST_PAINEL_COVIDBR_2021_Parte1_24jun2022.csv',sep=';', decimal=',')
#TABELA 2021 PARTE 2
tabela2021_p2 = pd.read_csv('HIST_PAINEL_COVIDBR_2021_Parte2_24jun2022.csv',sep=';', decimal=',')
#TABELA 2022 PARTE 1
tabela2022_p1 = pd.read_csv('HIST_PAINEL_COVIDBR_2022_Parte1_24jun2022.csv',sep=';', decimal=',')
```

# Total de casos de COVID nos estados ao longo dos anos:

# 2020:

```python
estados2020 = tabela2020_p1['estado'].value_counts()
estados2020_p2 = tabela2020_p2['estado'].value_counts()
estados2020 += estados2020_p2
estados2020[:].plot(figsize=(15, 10), kind='bar',color ='red')
```

# 2021:

```python
estados2021 = tabela2021_p1['estado'].value_counts()
estados2021_p2 = tabela2021_p2['estado'].value_counts()
estados2021 += estados2021_p2
estados2021[:].plot(figsize=(15, 10), kind='bar',color ='purple')
```

# 2022:

```python

estados2022 = tabela2022_p1['estado'].value_counts()
estados2022[:].plot(figsize=(15, 10), kind='bar',color ='green')
```

# Total:

```python
total = 0
total += estados2020
total += estados2021
total += estados2022
print(total)
total[:].plot(figsize=(15, 10), kind='bar',color ='yellow')
```
